```julia
using CSV
#using CurveFit
using Dates, DataFrames
#, Distributions, DSP
#using Gtk
#using LaTeXStrings
using NativeFileDialog
using Plots
using Printf
using StatsBase #, Statistics
#using Tk

# Widen screen for better viewing
display(HTML("<style>:root { --jp-notebook-max-width: 80% !important; }</style>"))


function process_word_0(word)
##########################################    
    Tp = parse(Int16, word[1:4]; base=2)
    M = parse(Int8, word[5]; base=2)
    T = parse(Int8, word[6]; base=2)
    F = parse(Int8, word[7]; base=2)
    C = parse(Int8, word[8]; base=2)
    Tn = parse(Int16, word[9:12]; base=2)
    
#    println("Transmission No. ",Tn)
#    println("System file word 0: Tp = ",Tp,"; M = ",M,"; T = ",T,"; F = ",F,"; C = ",C,"; Tn = ",Tn)
    
    return()
    
    end    # process_word_0()


function process_word_1(word)
##########################################    
    
    Hrms = parse(Int16, word[1:12]; base=2) / 400
    m0 = Hrms ^ 2
    Hm0 = 4 * sqrt(m0)
#    @printf("System file word 1: Hrms = %5.2fm; m0 = %5.5f; Hm0 = %5.2fm\n",Hrms,m0,Hm0)
    push!(hm0,Hm0)
    
    return()
    
    end    # process_word_1()


function process_word_2(word)
##########################################    
    
    fz = parse(Int16, word[5:12]; base=2) / 400
    Tz = 1/fz
#    @printf("System file word 2: fz = %5.4fHz; Tz = %5.2fs\n",fz,Tz)
    push!(tz,Tz)
    
    return()
    
    end    # process_word_2()


function process_word_3(word)
##########################################    
    
    PSDmax = 5000 * exp(-parse(Int32, word[1:12]; base=2) / 200)
#    @printf("System file word 3: PSDmax = %5.4fm¬≤/Hz\n",PSDmax)
    push!(max_PSD,PSDmax)

    return()
    
    end    # process_word_3()


function process_word_4(word)
##########################################    
    
    Tr = parse(Int16, word[3:12]; base=2) / 20 - 5
#    println("System file word 4: Tr = ",Tr,"·µíC")

    return()
    
    end    # process_word_4()


function process_word_5(word)
##########################################    
    
    Tw = parse(Int16, word[3:12]; base=2) / 20 - 5
#    println("System file word 5: Tw = ",Tw,"·µíC")
    push!(tw,Tw)
    
    return()
    
    end    # process_word_5()


function process_word_6(word)
##########################################    
    
    B = parse(Int16, word[10:12]; base=2)
    tol = parse(Int16, word[1:8]; base=2)
#    println("System file word 6: B = ",B,"; Tol = ",tol)

    return()
    
    end    # process_word_6()


function process_word_7(word)
##########################################    
    
    Av0 = parse(Int16, word[2:12]; base=2) / 800
    sign = parse(Int16, word[1]; base=2)
#==    
    if (sign==0)   
        println("System file word 7: Av0 = ",Av0,"m/s¬≤")
    elseif (sign==1)
        println("System file word 7: Av0 = ",-Av0,"m/s¬≤")
    end
==#    
    return(Av0)
    
    end    # process_word_7()


function process_word_8(word)
##########################################    
    
    Ax0 = parse(Int16, word[2:12]; base=2)
    sign = parse(Int16, word[1]; base=2)
#==    
    if (sign==0)   
        println("System file word 8: Ax0 = ",Ax0,"m/s¬≤")
    elseif (sign==1)
        println("System file word 8: Ax0 = ",-Ax0,"m/s¬≤")
    end
==#    
    return()
    
    end    # process_word_8()


function process_word_9(word)
##########################################    
    
    Ay0 = parse(Int16, word[2:12]; base=2)
    sign = parse(Int16, word[1]; base=2)
#==    
    if (sign==0)   
        println("System file word 9: Ay0 = ",Ay0,"m/s¬≤")
    elseif (sign==1)
        println("System file word 9: Ay0 = ",-Ay0,"m/s¬≤")
    end
==#    
    return()
    
    end    # process_word_9()


function process_word_10(word)
##########################################    
    
    LatMSB = word[2:12]
    SignLat = word[1]
    
##    println("System file word 10 done")
    
    return(LatMSB,SignLat)
    
    end    # process_word_10()


function process_word_11(word,LatMSB,SignLat)
##########################################    
    
    global LatLSB = word[1:12]
    
    lat = 90 * (parse(Int32, (LatMSB*LatLSB); base=2) / 2^23)
    if (SignLat=='0')   
#        @printf("System file words 10 and 11: Latitude = %5.4f·µíN\n",lat)
        push!(latitude,lat)
    elseif (SignLat=='1')
#        @printf("System file words 10 and 11: Latitude = %5.4f·µíS\n",lat)
        push!(latitude,-lat)        
    end
    
    return()
    
    end    # process_word_11()


function process_word_12(word)
##########################################    
    
    SignLon = word[1]
    LonMSB = word[2:12]
    
##    println("System file word 12 done")
    
    return(LonMSB,SignLon)
    
    end    # process_word_12()


function process_word_13(word,LonMSB,Sign)
##########################################    
    
    global LonLSB = word[1:12]
    
    lon = 180 * (parse(Int64, (LonMSB*LonLSB); base=2) / 2^23)
   
    if (Sign=='0')   
#        @printf("System file words 12 and 13: Longitude = %5.4f·µíE\n",lon)
        push!(longitude,lon)
        
    elseif (Sign=='1')
#        @printf("System file words 12 and 13: Longitude = %5.4f·µíW\n",lon)
        push!(longitude,-lon)
    end
    
    return()
    
    end    # process_word_13()


function process_word_14(word)
##########################################    
    
    O = 360 * (parse(Int16, word[5:12]; base=2) / 256)
    
#    println("System file word 14: O = ",O,"·µí")
    
    return()
    
    end    # process_word_14()


function process_word_15(word)
##########################################    
    
    IncMSB = parse(Int16, word[5:12]; base=2)
    IncLSB = parse(Int16, word[1:4]; base=2)
    
    I = (90/128) * (IncMSB - 128 + IncLSB/16)
#    @printf("System file word 15: I = %5.4f·µí\n\n",I)
    
    return()
    
    end    # process_word_15()


function calc_freq(freq_index)
##########################################    
    
    if freq_index < 16
        return(0.025 + freq_index*0.005)
    elseif freq_index > 16
        return(0.11+ (freq_index-16)*0.01)
    else
        return(0.10125)    # n = 15 is an intermediate point (see paragraph above table 5.7.3)
    end
    
end    # calc_freq()   


function process_spectrum_file(spectrum_file, aest)
##########################################    
    
    ll = length(spectrum_file)
    
    cc = 1
    global cyclic_data = string.(parse.(Int,"0x".*(spectrum_file)), base=2, pad=16);

#==
    for i in 1:ll

        println(cyclic_data[i])

    end
==#
    
    SùíçùíîùíÉ = []
    f‚Çô = Float32[]
    dir‚Çô = Float32[]

    N‚ÇÇùíçùíîùíÉ = []
    RPSD = Float32[]
    S = []
    m‚ÇÇ = Float32[]
    n‚ÇÇ = Float32[]
    K = Float32[]


    cc = 0

    # Calculate cyclic data block (see Datawell Waves5 Reference Manual 11.1.1.2, p.124)
    for i in 1:4:ll-3

        SùíçùíîùíÉ = parse(Int, SubString(cyclic_data[cc+i], 1, 2); base=2)
        push!(f‚Çô, calc_freq(parse(Int, SubString(cyclic_data[cc+i], 3, 8); base=2)))
        push!(dir‚Çô, parse(Int, SubString(cyclic_data[cc+i], 9, 16); base=2) * 360 / 256)

        M‚ÇÇùíçùíîùíÉ = parse(Int, SubString(cyclic_data[cc+i+1], 1, 2); base=2)
        N‚ÇÇùíçùíîùíÉ = parse(Int, SubString(cyclic_data[cc+i+1], 3, 4); base=2)
        push!(RPSD,exp(-parse(Int, SubString(cyclic_data[cc+i+1], 5, 16); base=2) / 200))

        push!(S, 0.4476 * (parse(Int, SubString(cyclic_data[cc+i+2], 1, 8); base=2) + SùíçùíîùíÉ/4))
        M‚ÇÇ = parse(Int, SubString(cyclic_data[cc+i+2], 9, 16); base=2)
        push!(m‚ÇÇ,(M‚ÇÇ + M‚ÇÇùíçùíîùíÉ/4 - 128) / 128)   
        N‚ÇÇ = parse(Int, SubString(cyclic_data[cc+i+3], 1, 8); base=2)
        push!(n‚ÇÇ,(N‚ÇÇ + N‚ÇÇùíçùíîùíÉ/4 - 128) / 128)
        push!(K, parse(Int, SubString(cyclic_data[cc+i+3], 9, 16); base=2) / 100)

    end
#==    
    p1 = plot(f‚Çô, dir‚Çô, ylim=(0,360), lw=:2, xticks = 0:0.05:0.64, yticks = 0:30:360, yflip = true, label="Direction", xlabel="Frequency (Hz)", ylabel="Direction", legend=:topleft)
    p1 = plot!(twinx(), f‚Çô, RPSD, lw=:2, color=:red, fillrange = 0, fillalpha = 0.05, fillcolor = :red, xticks=:none, ylim=[0,1], yticks = 0:0.1:1, ylabel="Spectral Density (m¬≤/Hertz)", label="Normalised PSD")

    p_all = plot(p1, framestyle = :box, xlim=[0,0.64], title="DATAWELL Spectra and Direction\n"*Dates.format(aest, "dd/mm/yy HH:MM"), titlefontsize=:10,
        leftmargin = 15Plots.mm, rightmargin = 25Plots.mm, bottommargin = 15Plots.mm, topmargin = 5Plots.mm, size=(1400, 600), colorbar=false, gridlinewidth=0.5, gridstyle=:dot, gridalpha=1, fg_legend = :false)

    display(p_all)
==#    
    
    return(f‚Çô, RPSD, dir‚Çô, S)
    
end    # process_spectrum_file()


#

#
```

```julia
using DataFrames
using EasyFit
using Peaks, Plots


# Function to find local maxima and minima
function find_local_extrema(RPSD)
##################################
    
    local_maxima = argmaxima(RPSD, 3; strict=true)
    local_minima = argminima(RPSD, 3; strict=true)
    
    return(local_maxima, local_minima)
    
end    # find_local_extrema()


# Calculate a PM spectrum for a given peak frequency
function pierson_moskowitz_spectrum(f, fp)
##########################################
    
    Œ± = 0.0081
    g = 9.81
    
    return(Œ± * g^2 * (2œÄ)^-4 * f .^ -4 .* exp.(-5/4 * (fp ./ f) .^ 4))
    
end    # pierson_moskowitz_spectrum()

# Check for significant low-frequency energy
function check_low_frequency_energy(RPSD, f2, threshold=0.02)
##############################################################
    
    low_freq_indices = findall(x -> x < threshold, f2)
    
    return(any(x -> x > 0.02, RPSD[low_freq_indices]))  # Set a significance threshold for low-frequency energy

end    # check_low_frequency_energy()


# Main analysis function
function analyse_spectra(df)
############################
    
    results = DataFrame(Date=[], peak_frequency=Float64[], peak_type=String[], second_peak_frequency=Float64[], 
        second_peak_value=Float64[], min_frequency=Float64[], min_value=Float64[], low_frequency_warning=Bool[])
    
    for i in 1:nrow(df)
        
        date = df.Date[i]
        RPSD = df.RPSD[i] .* df.PSDmax[i]
        f2 = df.Freq[i]
        
        # Find local maxima and minima
        local_maxima, local_minima = find_local_extrema(RPSD)
        
        # Identify the peak of the spectra
        peak_index = argmax(RPSD)
        peak_frequency = f2[peak_index]
        peak_value = RPSD[peak_index]
        
        # Calculate PM spectrum value at peak frequency
        pm_peak_value = pierson_moskowitz_spectrum(peak_frequency, peak_frequency)
        
        second_peak_frequency = NaN
        second_peak_value = NaN
        min_frequency = NaN
        min_value = NaN

##        println(date," ",peak_frequency," ",peak_value," ",pm_peak_value)
        
        if peak_value > pm_peak_value
            
            peak_type = "Wind Sea"
            # Analyse values to the left of the peak
            left_maxima = local_maxima[local_maxima .< peak_index]
            left_minima = local_minima[local_minima .< peak_index]
            
            if !isempty(left_maxima)
                second_peak_index = left_maxima[argmax(RPSD[left_maxima])]
                second_peak_frequency = f2[second_peak_index]
                second_peak_value = RPSD[second_peak_index]
                
                if !isempty(left_minima)
                    
                    min_between_peaks = left_minima[(left_minima .> second_peak_index) .& (left_minima .< peak_index)]
                    if !isempty(min_between_peaks)
                        min_index = min_between_peaks[argmin(RPSD[min_between_peaks])]
                        min_frequency = f2[min_index]
                        min_value = RPSD[min_index]
                    end
                    
                end
                
            end
            
        else
            
            peak_type = "Ocean Swell"
            # Analyise values to the right of the peak
            right_maxima = local_maxima[local_maxima .> peak_index]
            right_minima = local_minima[local_minima .> peak_index]
            
            if !isempty(right_maxima)
                
                second_peak_index = right_maxima[argmax(RPSD[right_maxima])]
                second_peak_frequency = f2[second_peak_index]
                second_peak_value = RPSD[second_peak_index]
                
                if !isempty(right_minima)
                    
                    min_between_peaks = right_minima[(right_minima .> peak_index) .& (right_minima .< second_peak_index)]
                    if !isempty(min_between_peaks)
                        min_index = min_between_peaks[argmin(RPSD[min_between_peaks])]
                        min_frequency = f2[min_index]
                        min_value = RPSD[min_index]
                    end
                    
                end
                
            end
            
        end
        
        # Check for low-frequency energy warning
        low_frequency_warning = check_low_frequency_energy(RPSD, f2)
        
        push!(results, (date, peak_frequency, peak_type, second_peak_frequency, second_peak_value, min_frequency, min_value, low_frequency_warning))
    end
    
    return results

end    # analyse_spectra()


# Function to calculate Hm0
function calculate_Hm0(frequencies, spectral_values)
###############################################
    
    # Trapezoidal integration to calculate the area under the spectral curve
    area = sum(0.5 * (frequencies[2:end] .- frequencies[1:end-1]) .* (spectral_values[2:end] .+ spectral_values[1:end-1]))

    # Calculate Hm0
    Hm0 = 4 * sqrt(area)
    
    return(Hm0)

end    # calc_Hmo()


# Function to calculate Tp
function calculate_Tp(frequencies, spectral_values)
###################################################
    
    # Find the peak frequency
    peak_index = argmax(spectral_values)
    peak_frequency = frequencies[peak_index]
    
    # Calculate Tp
    Tp = 1 / peak_frequency
    
    return(Tp)
    
end    # calculate_Tp()


########################################################################################################################
########################################################################################################################
########################################################################################################################
display(HTML("<style>.jp-Cell { width: 120% !important; }</style>"))

# Analyse the spectra and get results
results = analyse_spectra(SDT_df);

# Join the two DataFrames on the Date column
results = innerjoin(results, SDT_df, on=:Date)

### Calculate the Hm0 for both Sea and Swell
Hm0 = Float64[]; Hm0_swell = Float64[]; Hm0_wind_sea = Float64[]
Tp = Float64[]; Tp_swell = Float64[]; Tp_wind_sea = Float64[]

for row in eachrow(results)
    
    frequencies = row.Freq
    spectral_values = row.RPSD

    # calculate Hm0 and Tp over entire spectra
    push!(Hm0, calculate_Hm0(frequencies, spectral_values))
    push!(Tp, calculate_Tp(frequencies, spectral_values))
        
    sep_freq = row.min_frequency
    peak_type = row.peak_type
    
    if isnan(sep_freq)
        
        if peak_type == "Ocean Swell"
            # Entire spectrum is Ocean Swell
            push!(Hm0_swell, calculate_Hm0(frequencies, spectral_values))
            push!(Hm0_wind_sea,NaN)
        elseif peak_type == "Wind Sea"
            # Entire spectrum is Wind Sea
            push!(Hm0_swell,NaN)
            push!(Hm0_wind_sea,calculate_Hm0(frequencies, spectral_values))
        else
            push!(Hm0_swell,NaN)
            push!(Hm0_wind_sea,NaN)
        end
        
    else
        
        # Find the indices for Ocean Swell and Wind Sea
        swell_indices = findall(x -> x <= sep_freq, frequencies)
        wind_sea_indices = findall(x -> x > sep_freq, frequencies)
        
        # Separate the spectral values and frequencies
        swell_frequencies = frequencies[swell_indices]
        swell_spectral_values = spectral_values[swell_indices]
        wind_sea_frequencies = frequencies[wind_sea_indices]
        wind_sea_spectral_values = spectral_values[wind_sea_indices]
        
        # Calculate Hm0 and Tp for Ocean Swell and Wind Sea
        push!(Hm0_swell,calculate_Hm0(swell_frequencies,swell_spectral_values))
        push!(Hm0_wind_sea,calculate_Hm0(wind_sea_frequencies, wind_sea_spectral_values))
        
    end
    
end

# Extract peak_frequency values where peak_type is "Wind Sea"
peak_wind_sea_frequencies = results[results.peak_type .== "Wind Sea", :peak_frequency]
peak_ocean_swell_frequencies = results[results.peak_type .== "Ocean Swell", :peak_frequency]

# Extract second_peak_frequency values where peak_type is "Ocean Swell"
minor_ocean_swell_frequencies = results[results.peak_type .== "Ocean Swell", :second_peak_frequency]
minor_wind_sea_frequencies = results[results.peak_type .== "Wind Sea", :second_peak_frequency]

# Combine these frequencies into a single array
combined_wind_sea_frequencies = vcat(peak_wind_sea_frequencies, minor_ocean_swell_frequencies)
combined_ocean_swell_frequencies = vcat(peak_ocean_swell_frequencies, minor_wind_sea_frequencies)

Tp_wind_sea = 1 ./ combined_wind_sea_frequencies
Tp_swell = 1 ./ combined_ocean_swell_frequencies

# Do moving average filters over the Hm0 and Tp values
ma_filter = 3
ma_Hm0 = movavg(Hm0, ma_filter)
ma_wind_sea_Hm0 = movavg(Hm0_wind_sea, ma_filter)
ma_ocean_swell_Hm0 = movavg(Hm0_swell, ma_filter)

ma_Tp = movavg(Tp, ma_filter)
ma_wind_sea_Tp = movavg(Tp_wind_sea, ma_filter)
ma_ocean_swell_Tp = movavg(Tp_swell, ma_filter)

title = split(split(infil,"\\")[3]," ")[1]*" "*Dates.format(results.Date[1], "yyyy/mm/dd")*" to "*Dates.format(results.Date[end], "yyyy/mm/dd")
plot_file = ".\\Plots\\"*replace(replace(title, "/" => "_"), " " => "_")*"_SDT_sea_swell_split.png"

# Plot the Hm0 values
p1 = Plots.plot(results.Date, ma_Hm0.x, label="Combined Hm0", lc=:green, lw=:3, alpha=0.25, yaxis="Hm0 (m)", title=title)
p1 = Plots.plot!(results.Date, ma_wind_sea_Hm0.x, label="Wind Sea Hm0", lc=:blue, lw=:3, alpha=0.5)
p1 = Plots.plot!(results.Date, ma_ocean_swell_Hm0.x, label="Ocean Swell Hm0", lc=:red, lw=:3, alpha=0.5)

# Plot the Tp values
p2 = Plots.plot(results.Date, ma_Tp.x, label="Combined Tp", lc=:green, lw=:3, alpha=0.25, yaxis="Tp (s)")
p2 = Plots.plot!(results.Date, ma_wind_sea_Tp.x, label="Wind Sea Tp", lc=:blue, lw=:3, alpha=0.5)
p2 = Plots.plot!(results.Date, ma_ocean_swell_Tp.x, label="Ocean Swell Tp", lc=:red, lw=:3, alpha=0.5)

#tm_tick = range(SDT_df.Date[1],SDT_df.Date[end],step=Day(1))
tm_tick = range(DateTime(Date(results.Date[1])), DateTime(Date(results.Date[end] + Day(1))),step=Day(1))
ticks = Dates.format.(tm_tick,"dd")

plot_p1_p2 = Plots.plot(p1, p2, layout=(2,1), xlims=(DateTime(Date(results.Date[1])), DateTime(Date(results.Date[end] + Day(1)))), xticks=(tm_tick,ticks), xtickfontsize=8, xaxis="Date", 
    ylim=(0,Inf), ytickfontsize=8, titlefontsize=12, framestyle=:box, legendfontsize=12, legend=:topright,
    fg_legend=:transparent, bg_legend=:transparent, plot_padding=1Plots.mm,
    leftmargin = 15Plots.mm, bottommargin = 5Plots.mm, grid=true, size=(1600,800), gridlinewidth=0.5, gridstyle=:dot, gridalpha=1)

try
                                                
    savefig(plot_file)
    println("\nPlot file saved as "*plot_file)

catch

    "Alert: Plot not saved!"

end

display(plot_p1_p2)
```

```julia
# Function to find the peak spectral value within a specified frequency range
function find_peak_in_range(df, freq_min, freq_max)
###################################################
    
    peak_value = -Inf
    peak_row_index = -1

    for (i, row) in enumerate(eachrow(df))
        frequencies = row.f2
        spectral_values = row.Pden2

        # Find indices within the specified frequency range
        indices_in_range = findall(freq_min .<= frequencies .<= freq_max)
        
        if !isempty(indices_in_range)
            # Get spectral values within the specified range
            spectral_values_in_range = spectral_values[indices_in_range]

            # Find the peak spectral value in the range
            max_value_in_range = maximum(spectral_values_in_range)

            # Update the peak value and row index if a new peak is found
            if max_value_in_range > peak_value
                peak_value = max_value_in_range
                peak_row_index = i
            end
        end
    end

    return(peak_row_index, peak_value)
    
end    # find_peak_in_range()


# Function to find the top N peak spectral values within a specified frequency range
function find_top_peaks_in_range(df, freq_min, freq_max, top_n)
###############################################################
    
    peaks = []

    for (i, row) in enumerate(eachrow(df))
        frequencies = row.Freq
        spectral_values = row.RPSD .* row.PSDmax

        # Find indices within the specified frequency range
        indices_in_range = findall(freq_min .<= frequencies .<= freq_max)
        
        if !isempty(indices_in_range)
            # Get spectral values within the specified range
            spectral_values_in_range = spectral_values[indices_in_range]

            # Find the peak spectral value in the range
            max_value_in_range = maximum(spectral_values_in_range)

            # Append the peak value and row index to the peaks array
            push!(peaks, (max_value_in_range, i))
        end
    end
    
# Sort peaks by the spectral value in descending order and take the top N
    sorted_peaks = sort(peaks, by = x -> -x[1])
    top_peaks = sorted_peaks[1:min(top_n, length(sorted_peaks))]

    return(top_peaks)
    
end    # find_top_peaks_in_range()


# Function to convert frequency in Hertz to Period in Seconds
function convert_frequency_to_period(frequencies)
#################################################
    
    return(1.0 ./ frequencies)
    
end    # convert_frequency_to_period()


#######################################################################################################
#######################################################################################################
#######################################################################################################

# Define the frequency range
freq_min = 0.0
freq_max = 0.05

# Find the row with the peak spectral value in the specified frequency range
##peak_row_index, peak_value = find_peak_in_range(filtered_df, freq_min, freq_max)

# Number of top peaks to find
top_n = 10

# Find the top N rows with the peak spectral values in the specified frequency range
top_peaks = find_top_peaks_in_range(filtered_df, freq_min, freq_max, top_n)

#==
# Display the result
if peak_row_index != -1
    println("Row with peak spectral value within the range $freq_min to $freq_max Hz:")
    println(filtered_df[peak_row_index, :])
    println("Peak spectral value: $peak_value")
else
    println("No peak found within the range $freq_min to $freq_max Hz.")
end
==#

p1 = Plots.plot()
freq_range = 1:6
# Plotting each row
for row in eachrow(filtered_df)
    frequencies_hz = row.Freq[freq_range]
    spectral_values = row.RPSD[freq_range] .* row.PSDmax
    periods_sec = convert_frequency_to_period(frequencies_hz)
    
    # Create plot for the current row
    p1 = Plots.plot!(periods_sec, spectral_values, label="", lw=:2, lc=:grey, alpha=0.5,)
end


for (_, ii) in top_peaks
    frequencies_hz = filtered_df.Freq[ii][freq_range]
    spectral_values = filtered_df.RPSD[ii][freq_range] .* filtered_df.PSDmax[ii]
    periods_sec = convert_frequency_to_period(frequencies_hz)
    peak_time_string = Dates.format(filtered_df.Date[ii], "yyyy-mm-dd HH:MM")
    p1 = Plots.plot!(periods_sec, spectral_values, label=peak_time_string, lw=:3, alpha=0.5,)
end

#Dates.format(start_date, "yyyy-mm-dd HH:MM:SS")
title = split(split(infil,"\\")[3]," ")[1]*" "*Dates.format(filtered_df.Date[1], "yyyy/mm/dd")*" to "*Dates.format(filtered_df.Date[end], "yyyy/mm/dd HH:MM")
plot_file = ".\\Plots\\"*replace(replace(title, "/" => "_"), " " => "_")*"_SDT_top10_sea_swell_split.png"

p1_plot = plot(p1, xtickfontsize=7, title=title,
    ytickfontsize=8, yaxis=:log, xlims=(20,45), xminorticks=5, legend=:bottomright,
    fg_legend=:transparent, bg_legend=:transparent, plot_padding=1Plots.mm,
    framestyle = :box, xlabel="Period (Seconds)", ylabel="Spectral Values",
    leftmargin = 15Plots.mm, bottommargin = 15Plots.mm, grid=true, size=(1800,800), gridlinewidth=0.5, gridstyle=:dot, gridalpha=1)

try
                                                
    savefig(plot_file)
    println("\nPlot file saved as "*plot_file)

catch

    "Alert: Plot not saved!"

end

display(p1_plot)
```

```julia
using DataFrames
using Statistics
using FFTW

# Define a function to calculate MAD for a 1D array
function mad(arr)
    med = median(arr)
    mad_value = median(abs.(arr .- med))
    return mad_value
end

# Define a function to calculate MAD for a matrix
function mad_matrix(matrix; dims=1)
    if dims == 1
        return [mad(matrix[:, i]) for i in 1:size(matrix, 2)]
    elseif dims == 2
        return [mad(matrix[i, :]) for i in 1:size(matrix, 1)]
    else
        error("Unsupported dimension: $dims")
    end
end

# Extract all spectral arrays from the DataFrame
spectral_values = SDT_noise.RPSD .* SDT_noise.PSDmax

# Convert the list of arrays into a matrix where each row is a spectrum
spectral_matrix = hcat(spectral_values...)'

# Calculate the mean spectra (mean of each column) directly
mean_spectra = mean(spectral_matrix, dims=1)
mean_spectra_vector = vec(mean_spectra)

# Calculate the median spectra (median of each column)
median_spectra = median(spectral_matrix, dims=1)

# Calculate the MAD (median absolute deviation) for each frequency
mad_spectra = mad_matrix(spectral_matrix, dims=1)

# Convert the SDT_df to vectors
median_spectra_vector = vec(median_spectra)
mad_spectra_vector = vec(mad_spectra)

println("Mean Spectra: ", mean_spectra_vector)
println("Median Spectra: ", median_spectra_vector)
println("MAD Spectra: ", mad_spectra_vector)

# Define a function to remove outliers using MAD
function remove_outliers_using_mad(matrix, threshold=3.0)
    median_vals = median(matrix, dims=1)
    mad_vals = mad_matrix(matrix, dims=1)
    lower_bound = median_vals .- threshold .* mad_vals
    upper_bound = median_vals .+ threshold .* mad_vals
    filtered_matrix = copy(matrix)
    
    for i in 1:size(matrix, 2)  # iterate over columns
        for j in 1:size(matrix, 1)  # iterate over rows
            if matrix[j, i] < lower_bound[i] || matrix[j, i] > upper_bound[i]
                filtered_matrix[j, i] = median_vals[i]
            end
        end
    end
    
    return filtered_matrix
end

# Remove outliers from the spectral matrix
filtered_spectral_matrix = remove_outliers_using_mad(spectral_matrix)

# Calculate the mean spectra after outlier removal
mean_spectra_filtered = mean(filtered_spectral_matrix, dims=1)
mean_spectra_vector_filtered = vec(mean_spectra_filtered)

println("Filtered Mean Spectra: ", mean_spectra_vector_filtered)

# Compute the difference spectra
difference_spectra = mean_spectra_vector - mean_spectra_vector_filtered

println("Difference Spectra: ", difference_spectra)

# Perform the inverse Fourier transform to get the time series of the outliers
time_series_outliers = real(ifft(difference_spectra))

println("Time Series of Outliers: ", time_series_outliers)
```

```julia
using CSV
#using CurveFit
using Dates, DataFrames
#, Distributions, DSP
#using Gtk
#using LaTeXStrings
using NativeFileDialog
using Plots
using Printf
using StatsBase #, Statistics
#using Tk


function file_search(search_valu)
    # select directory to start file search
    parent_directory = pick_folder()
    println("Selected ",parent_directory)

    # Search for all files that match search value
    rootpath = parent_directory
    pattern  = search_valu #".SDT"

    files_list = []
    for (root, dirs, files) in walkdir(rootpath)
        for file in files

            if occursin(pattern, file) push!(files_list, root*"\\"*file) end
        end
    end

    return(files_list)
    
    end    # file_search()


function process_word_0(word)
##########################################    
    Tp = parse(Int16, word[1:4]; base=2)
    M = parse(Int8, word[5]; base=2)
    T = parse(Int8, word[6]; base=2)
    F = parse(Int8, word[7]; base=2)
    C = parse(Int8, word[8]; base=2)
    Tn = parse(Int16, word[9:12]; base=2)
    
#    println("Transmission No. ",Tn)
#    println("System file word 0: Tp = ",Tp,"; M = ",M,"; T = ",T,"; F = ",F,"; C = ",C,"; Tn = ",Tn)
    
    return()
    
    end    # process_word_0()


function process_word_1(word)
##########################################    
    
    Hrms = parse(Int16, word[1:12]; base=2) / 400
    m0 = Hrms ^ 2
    Hm0 = 4 * sqrt(m0)
#    @printf("System file word 1: Hrms = %5.2fm; m0 = %5.5f; Hm0 = %5.2fm\n",Hrms,m0,Hm0)
    push!(hm0,Hm0)
    
    return()
    
    end    # process_word_1()


function process_word_2(word)
##########################################    
    
    fz = parse(Int16, word[5:12]; base=2) / 400
    Tz = 1/fz
#    @printf("System file word 2: fz = %5.4fHz; Tz = %5.2fs\n",fz,Tz)
    push!(tz,Tz)
    
    return()
    
    end    # process_word_2()


function process_word_3(word)
##########################################    
    
    PSDmax = 5000 * exp(-parse(Int32, word[1:12]; base=2) / 200)
#    @printf("System file word 3: PSDmax = %5.4fm¬≤/Hz\n",PSDmax)
    push!(max_PSD,PSDmax)

    return()
    
    end    # process_word_3()


function process_word_4(word)
##########################################    
    
    Tr = parse(Int16, word[3:12]; base=2) / 20 - 5
#    println("System file word 4: Tr = ",Tr,"·µíC")

    return()
    
    end    # process_word_4()


function process_word_5(word)
##########################################    
    
    Tw = parse(Int16, word[3:12]; base=2) / 20 - 5
#    println("System file word 5: Tw = ",Tw,"·µíC")
    push!(tw,Tw)
    
    return()
    
    end    # process_word_5()


function process_word_6(word)
##########################################    
    
    B = parse(Int16, word[10:12]; base=2)
    tol = parse(Int16, word[1:8]; base=2)
#    println("System file word 6: B = ",B,"; Tol = ",tol)

    return()
    
    end    # process_word_6()


function process_word_7(word)
##########################################    
    
    Av0 = parse(Int16, word[2:12]; base=2) / 800
    sign = parse(Int16, word[1]; base=2)
#==    
    if (sign==0)   
        println("System file word 7: Av0 = ",Av0,"m/s¬≤")
    elseif (sign==1)
        println("System file word 7: Av0 = ",-Av0,"m/s¬≤")
    end
==#    
    return()
    
    end    # process_word_7()


function process_word_8(word)
##########################################    
    
    Ax0 = parse(Int16, word[2:12]; base=2)
    sign = parse(Int16, word[1]; base=2)
#==    
    if (sign==0)   
        println("System file word 8: Ax0 = ",Ax0,"m/s¬≤")
    elseif (sign==1)
        println("System file word 8: Ax0 = ",-Ax0,"m/s¬≤")
    end
==#    
    return()
    
    end    # process_word_8()


function process_word_9(word)
##########################################    
    
    Ay0 = parse(Int16, word[2:12]; base=2)
    sign = parse(Int16, word[1]; base=2)
#==    
    if (sign==0)   
        println("System file word 9: Ay0 = ",Ay0,"m/s¬≤")
    elseif (sign==1)
        println("System file word 9: Ay0 = ",-Ay0,"m/s¬≤")
    end
==#    
    return()
    
    end    # process_word_9()


function process_word_10(word)
##########################################    
    
    LatMSB = word[2:12]
    SignLat = word[1]
    
##    println("System file word 10 done")
    
    return(LatMSB,SignLat)
    
    end    # process_word_10()


function process_word_11(word,LatMSB,SignLat)
##########################################    
    
    global LatLSB = word[1:12]
    
    lat = 90 * (parse(Int32, (LatMSB*LatLSB); base=2) / 2^23)
    if (SignLat=='0')   
#        @printf("System file words 10 and 11: Latitude = %5.4f·µíN\n",lat)
        push!(latitude,lat)
    elseif (SignLat=='1')
#        @printf("System file words 10 and 11: Latitude = %5.4f·µíS\n",lat)
        push!(latitude,-lat)        
    end
    
    return()
    
    end    # process_word_11()


function process_word_12(word)
##########################################    
    
    SignLon = word[1]
    LonMSB = word[2:12]
    
##    println("System file word 12 done")
    
    return(LonMSB,SignLon)
    
    end    # process_word_12()


function process_word_13(word,LonMSB,Sign)
##########################################    
    
    global LonLSB = word[1:12]
    
    lon = 180 * (parse(Int64, (LonMSB*LonLSB); base=2) / 2^23)
   
    if (Sign=='0')   
#        @printf("System file words 12 and 13: Longitude = %5.4f·µíE\n",lon)
        push!(longitude,lon)
        
    elseif (Sign=='1')
#        @printf("System file words 12 and 13: Longitude = %5.4f·µíW\n",lon)
        push!(longitude,-lon)
    end
    
    return()
    
    end    # process_word_13()


function process_word_14(word)
##########################################    
    
    O = 360 * (parse(Int16, word[5:12]; base=2) / 256)
    
#    println("System file word 14: O = ",O,"·µí")
    
    return()
    
    end    # process_word_14()


function process_word_15(word)
##########################################    
    
    IncMSB = parse(Int16, word[5:12]; base=2)
    IncLSB = parse(Int16, word[1:4]; base=2)
    
    I = (90/128) * (IncMSB - 128 + IncLSB/16)
#    @printf("System file word 15: I = %5.4f·µí\n\n",I)
    
    return()
    
    end    # process_word_15()


function calc_freq(freq_index)
##########################################    
    
    if freq_index < 16
        return(0.025 + freq_index*0.005)
    elseif freq_index > 16
        return(0.11+ (freq_index-16)*0.01)
    else
        return(0.10125)    # n = 15 is an intermediate point (see paragraph above table 5.7.3)
    end
    
end    # calc_freq()   


function process_spectrum_file(spectrum_file, aest)
##########################################    
    
    ll = length(spectrum_file)
    
    cc = 1
    global cyclic_data = string.(parse.(Int,"0x".*(spectrum_file)), base=2, pad=16);

#==
    for i in 1:ll

        println(cyclic_data[i])

    end
==#
    
    SùíçùíîùíÉ = []
    f‚Çô = Float32[]
    dir‚Çô = Float32[]

    N2ùíçùíîùíÉ = []
    RPSD = Float32[]
    S = []
    m2 = Float32[]
    n2 = Float32[]
    K = Float32[]


    cc = 0

    # Calculate cyclic data block (see Datawell Waves5 Reference Manual 11.1.1.2, p.124)
    for i in 1:4:ll-3

        SùíçùíîùíÉ = parse(Int, SubString(cyclic_data[cc+i], 1, 2); base=2)
        push!(f‚Çô, calc_freq(parse(Int, SubString(cyclic_data[cc+i], 3, 8); base=2)))
        push!(dir‚Çô, parse(Int, SubString(cyclic_data[cc+i], 9, 16); base=2) * 360 / 256)

        m2ùíçùíîùíÉ = parse(Int, SubString(cyclic_data[cc+i+1], 1, 2); base=2)
        N2ùíçùíîùíÉ = parse(Int, SubString(cyclic_data[cc+i+1], 3, 4); base=2)
        push!(RPSD,exp(-parse(Int, SubString(cyclic_data[cc+i+1], 5, 16); base=2) / 200))

        push!(S, 0.4476 * (parse(Int, SubString(cyclic_data[cc+i+2], 1, 8); base=2) + SùíçùíîùíÉ/4))
        m2 = parse(Int, SubString(cyclic_data[cc+i+2], 9, 16); base=2)
        push!(m2,(m2 + m2ùíçùíîùíÉ/4 - 128) / 128)   
        n2 = parse(Int, SubString(cyclic_data[cc+i+3], 1, 8); base=2)
        push!(n2,(n2 + N2ùíçùíîùíÉ/4 - 128) / 128)
        push!(K, parse(Int, SubString(cyclic_data[cc+i+3], 9, 16); base=2) / 100)

    end
    
    return(f‚Çô, RPSD, dir‚Çô, S)
    
end    # process_spectrum_file()


##########################################    
##########################################    
##########################################
###           Start of Main            ###
##########################################    
##########################################    
##########################################    

# select directory
##sdt_directory = pick_folder()

# build list of all csv files in selected directory
##sdt_files = filter(x->occursin(".SDT",x), readdir(sdt_directory));
sdt_files = file_search(".SDT")
# Create empty df to hold the spectral data
SDT_df = DataFrame([[],[],[],[],[],[],[],[],[],[]], ["Date", "UTC", "Hm0", "Tz", "SST", "PSDmax", "Freq", "RPSD", "Dirn", "Spread"])

for kk in sdt_files

    infil = kk
    #Change the type-interpretation of the binary file data from hex to string
    println("Reading BINARY data from ",kk)
    data_array = string.(read(infil), base=16, pad=2);

    date_time_list = []

    ii = 1
    #==
    # Convert df column types from 'Any' to their proper type
    SDT_df.Date = map(DateTime, SDT_df.Date);
    SDT_df.UTC = map(DateTime, SDT_df.UTC);
    SDT_df.Spectra = map(Float64, SDT_df.Spectra);
    SDT_df.System = map(Float64, SDT_df.System);
    ==#

##    println("Decoding SDT data now")
##    flush(stdout)

    # loop through selected SDT file until end
    while ii < length(data_array)

    ##    println("Start of record: ",ii)

        # get header info
        start_of_message = data_array[ii]
        message_id = data_array[ii+1]
        message_length = parse(Int,data_array[ii+2] * data_array[ii+3], base= 16)
        check_sum1 = data_array[ii+4]  

        yr = parse(Int,data_array[ii+5] * data_array[ii+6], base= 16)
        month = parse(Int, data_array[ii+7], base= 16)
        day = parse(Int, data_array[ii+8], base= 16)
        hour = parse(Int, data_array[ii+9], base= 16)
        minute = parse(Int, data_array[ii+10], base= 16)

        utc = DateTime(yr,month,day,hour,minute)
                   
        global aest = utc + Hour(10)
    ##    push!(date_time_list,aest)

        global spectrum_file = []
        global system_file = []

        # Process the spectrum file info
        for jj in (11:2:521) .+ ii

            push!(spectrum_file,data_array[jj] * data_array[jj+1])

        end

    ##    println("Spectrum file length: ",length(spectrum_file)) 

        # Process the system file info
        for jj in (523:2:554) .+ ii 

            push!(system_file, data_array[jj] * data_array[jj+1])

        end

    ##    println("System file length: ",length(system_file))

        freq, rpsd, dirn, spread = process_spectrum_file(spectrum_file, aest)

        global hm0 = []
        global tz = []
        global max_PSD = []
        global tw = []
        global latitude = []
        global longitude = []

        word = string.(parse.(Int,"0x".*system_file), base=2, pad=16)
        global Word_No = parse.(Int64, SubString.(word,1,4), base=2)
        global system_word = SubString.(word,5,16)    

        for i in eachindex(Word_No)

            if (i==1)
                process_word_0(system_word[i])
            elseif (i==2)
                process_word_1(system_word[i])
            elseif (i==3)
                process_word_2(system_word[i])
            elseif (i==4)
                process_word_3(system_word[i])
            elseif (i==5)
                process_word_4(system_word[i])
            elseif (i==6)
                process_word_5(system_word[i])
            elseif (i==7)
                process_word_6(system_word[i])
            elseif (i==8)
                process_word_7(system_word[i])
            elseif (i==9)
                process_word_8(system_word[i])
            elseif (i==10)
                process_word_9(system_word[i])
            elseif (i==11)
                global LatMSB,SignLat = process_word_10(system_word[i])
            elseif (i==12)
                try
                    process_word_11(system_word[i],LatMSB,SignLat)
                catch
                    println("Attempted to process System file word 11 - but no word 10 available")
                end
            elseif (i==13)
                global LonMSB,SignLon = process_word_12(system_word[i])
            elseif (i==14)
                try
                    process_word_13(system_word[i],LonMSB,SignLon)
                catch
                    println("Attempted to process System file word 13 - but no word 12 available")
                end
            elseif (i==15)
                process_word_14(system_word[i])
            elseif (i==16)
                process_word_15(system_word[i])
            else
                println("Error - Invalid System file word")       
            end

        end

        ii += message_length + 6

        push!(SDT_df,(utc .+ Hour(10), utc, mode(hm0), mode(tz), mode(tw), mode(max_PSD), freq, rpsd, dirn, spread))

    end

end
    
sort!(SDT_df, (:Date));
##########################################    
##########################################    
##########################################
###            End of Main             ###
##########################################    
##########################################    
##########################################    
```
